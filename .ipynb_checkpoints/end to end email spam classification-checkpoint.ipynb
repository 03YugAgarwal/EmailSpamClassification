# Email Spam Classification


### Problem Definition:
This is an end to end Email Spam Classification project where based on words present in email, we classify whether it is spam or not.

### Data Collection:
The data is available here: https://www.kaggle.com/datasets/balaka18/email-spam-classification-dataset-csv

### Evaluation:
If we can predict email spam with upto 80% accuracy, this project will be considered as success.

```python
plt.figure(figsize=(10,6))
plt.scatter(data.the[data.Prediction==1],data.to[data.Prediction==1],color=["salmon"])
plt.scatter(data.the[data.Prediction==0],data.to[data.Prediction==0],color=["lightblue"])
plt.xlabel('the')
plt.ylabel('to')
plt.title("the vs to")
plt.legend(["Spam","Not Spam"])
```
